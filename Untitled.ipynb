```python
from wordcloud import WordCloud
```

```python
f = open(u"1.txt", 'r').read()
print(f)
wordcloud = WordCloud(background_color="white",width=3000, height=1500, margin=2, font_path="Roboto-Regular.ttf").generate(f)
```

```python
import matplotlib.pyplot as plt
plt.imshow(wordcloud)
plt.axis("off")
plt.show()

wordcloud.to_file('test.png')
```
